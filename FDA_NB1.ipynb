```python
# Splitting the dose_vbm column into 3 seperate columns  :  dose_amt, dose_unit, dose_freq
import re

# Function to separate dose amount, unit, and frequency
def separate_dose(row):
   
    dose_frequency_pattern = r'(1X|Once|BID|BIW|HS|PRN|Q12H|Q2H|Q3H|Q3W|Q4H|Q5H|Q6H|Q8H|QD|QH|QID|QM|QOD|QOW|QW|TID|TIW|Daily|q2wk)'
    
    
    dose_amount_pattern = r'(\d+)'  # Matches number (e.g., '1000')
    dose_unit_pattern = r'(ug|mg|g|milligram|ml|milliliter|meq|DF|tabs|tablet|capsule|puff|IU|international unit|cp|packet)'  # Matches unit
    
    
    dose_amt_match = re.search(dose_amount_pattern, row)
    dose_amt = dose_amt_match.group(0) if dose_amt_match else None
    
    
    dose_unit_match = re.search(dose_unit_pattern, row, re.I)
    dose_unit = dose_unit_match.group(0) if dose_unit_match else None
    
   
    dose_freq_match = re.search(dose_frequency_pattern, row, re.I)
    dose_freq = dose_freq_match.group(0) if dose_freq_match else None

    return pd.Series([dose_amt, dose_unit, dose_freq])


df[['dose_amt', 'dose_unit', 'dose_freq']] = df['dose_vbm'].apply(separate_dose)


```

```python
# Converting all the drug_units to mg
conversion_dict = {
    'mg': 1, 'milligram': 1, 'g': 1000, 'milliliter': 1000, 'ml': 1000, 'ug': 0.001, 
    'MG': 1, 'mL': 1000, 'IU': 0.025, 'G': 1000, 'international unit': 0.025, 'iU': 0.025, 
    'Milligram': 1, 'Mg': 1, 'ML': 1000, 'UG': 0.001, 'MILLIGRAM': 1
     }


df['standardized_amt'] = df.apply(lambda row: row['dose_amt'] * conversion_dict.get(row['dose_unit'], 1), axis = 1)
```
